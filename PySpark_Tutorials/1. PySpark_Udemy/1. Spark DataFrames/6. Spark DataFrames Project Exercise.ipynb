## Bonus Question!
#### There are too many decimal places for mean and stddev in the describe() dataframe. Format the numbers to just show up to two decimal places. Pay careful attention to the datatypes that .describe() returns, we didn't cover how to do this exact formatting, but we covered something very similar. If you get stuck on this, don't worry, just view the solutions.

```python
df.describe().printSchema()
```

```python
# All string, so we need to convert them into float atfirst
```

```python
from pyspark.sql.functions import format_number
```

```python
result = df.describe()
result.show()
```

```python
result = df.describe()
result.select(result['summary'],
             format_number(result['Open'].cast('float'),2).alias('Open'),
             format_number(result['High'].cast('float'),2).alias('High'),
             format_number(result['Low'].cast('float'),2).alias('Low'),
             format_number(result['Close'].cast('float'),2).alias('Close'),
             result['Volume'].cast('int').alias('Volume'),
             format_number(result['Adj Close'].cast('float'),2).alias('Adj Close'),).show()
```

#### What is the max High per year?

```python
from pyspark.sql.functions import year
```

```python
df2 = df.withColumn('Year', year(df['Date']))
```

```python
df2.groupby('Year').agg({'High':'max'}).show()
```

```python
df2.groupby('Year').agg(max(df['High'])).show()
```

#### What is the average Close for each Calendar Month?
#### In other words, across all the years, what is the average Close price for Jan,Feb, Mar, etc... Your result will have a value for each of these months. 

```python
from pyspark.sql.functions import month, mean, avg
```

```python
df3 = df.withColumn('Month',month(df['Date']))
df3.collect()[0]
```

```python
df3.groupBy('Month').agg({'Close':'mean'}).show()
```

```python
df3.groupby('month').agg(mean(df['Close'])).show()
```

```python
df3.groupBy('Month').agg({'Close':'avg'}).show()
```

```python
df3.groupBy('Month').agg(mean(df['Close']), max(df['Close'])).show()
```

```python
df3.groupBy('Month').agg(avg(df['Close'])).orderBy('Month').show()
```

```python
df3.groupBy('Month').agg(avg(df['Close'])).orderBy(df3['Month']).show()
```

```python
df4 = df3.groupBy('Month').agg(avg(df['Close']))
df4.orderBy(df4['Month'].desc()).show()
```
